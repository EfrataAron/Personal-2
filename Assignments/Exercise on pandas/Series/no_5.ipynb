```python
import pandas as pd

# Load the dataset using a relative path
file_path = 'AfricaCupofNationsMatches.csv'
df = pd.read_csv(file_path)

# Display the original DataFrame
print("Original DataFrame:")
print(df.head())

# Find the most frequent value in the 'HomeTeamGoals' column
most_frequent_value = df['HomeTeamGoals'].mode()[0]
print(f"\nMost frequent value in 'HomeTeamGoals': {most_frequent_value}")

# Replace all other values with 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].apply(lambda x: x if x == most_frequent_value else 'Other')

# Display the modified DataFrame
print("\nModified DataFrame:")
print(df.head())

# Optionally, save the modified DataFrame to a new CSV file
output_file_path = 'Modified_AfricaCupofNationsMatches.csv'
df.to_csv(output_file_path, index=False)
print(f"\nModified DataFrame saved to {output_file_path}")
```
