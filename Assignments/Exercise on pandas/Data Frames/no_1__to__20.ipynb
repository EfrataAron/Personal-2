```python
# 3  Select specific columns
selected_columns = df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]
print("Selected columns:")
print(selected_columns.head())

# %pip install pandas
```

```python
# select the rows where Egypt appears.
# 4
# Select rows where Egypt appears
egypt_rows = df[(df['HomeTeam'] == 'Egypt') | (df['AwayTeam'] == 'Egypt')]
print("Rows where Egypt appears:")
print(egypt_rows)
print(egypt_rows.count())

# df=['AwayTeam']=df['AwayTeam'].str.strip()
# ethiopia_rows = df[(df['HomeTeam'] == 'Ethiopia') | (df['AwayTeam'] == 'Ethiopia')]
```

```python
# 5
#5.to count the number of rows and columns of your data frame.

# Count the number of rows and columns
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# 6
# 6.to select the rows where the ‘Attendance’ is missing.

# Select rows where 'Attendance' is missing
missing_attendance_rows = df[df['Attendance'].isna()]
print("Rows where 'Attendance' is missing:")
print(missing_attendance_rows)

```

```python
# 7
# 7.to select the rows where the ‘HomeTeamGoals’ are between 3 and 6 inclusive.

# Select rows where 'HomeTeamGoals' are between 3 and 6 inclusive
home_goals_between_3_and_6 = df[df['HomeTeamGoals'].between(3, 6)]
print("Rows where 'HomeTeamGoals' are between 3 and 6 inclusive:")
print(home_goals_between_3_and_6)
```

```python
# 8
# 8.to change the “AwayTeamGoals” in the 3rd row to 10.

# Change the 'AwayTeamGoals' in the 3rd row to 10
df.at[2, 'AwayTeamGoals'] = 10
print("Updated 3rd row 'AwayTeamGoals' to 10:")
print(df.iloc[2])
```

```python
# 9
# to sort the DataFrame first by ‘HomeTeam’ in ascending order,
# then by ‘HomeTeamScores’ in descending order.

# Sort the DataFrame by 'HomeTeam' in ascending order and 'HomeTeamGoals' in descending order
sorted_df = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])
print("DataFrame sorted by 'HomeTeam' and 'HomeTeamGoals':")
print(sorted_df.head())
```

```python
# 10
# to get list from DataFrame column headers

# Get list of column headers
column_headers = df.columns.tolist()
print("List of DataFrame column headers:")
print(column_headers)
```

```python
# 11
# to append a column of your choice to your DataFrame.

# Append a column 'MatchType' to the DataFrame
df['MatchType'] = 'Group Stage'
print("DataFrame with appended 'MatchType' column:")
print()
print(df.head())
```

```python
# 12
# 12. to add 2 rows to your DataFrame.

# Add 2 new rows to the DataFrame
new_rows = pd.DataFrame({
    'HomeTeam': ['TeamA', 'TeamB'],
    'AwayTeam': ['TeamC', 'TeamD'],
    'HomeTeamGoals': [2, 3],
    'AwayTeamGoals': [1, 2],
    'Attendance': [15000, 16000],
    'MatchType': ['Group Stage', 'Group Stage']
})
df = pd.concat([df, new_rows], ignore_index=True)
print("DataFrame with 2 new rows added:")
print()
print(df.tail())
```

```python
# 13
# 13.to change the country ‘Uganda’ to ‘China’ in the ‘AwayTeam’ column of the DataFrame.

# Change 'Uganda' to 'China' in the 'AwayTeam' column
df['AwayTeam'] = df['AwayTeam'].replace('Uganda', 'China')
print("DataFrame after changing 'Uganda' to 'China' in 'AwayTeam' column:")
print(df[df['AwayTeam'] == 'China'])
```

```python
# 14
# to reset index in your DataFrame.

# Reset the index of the DataFrame
df_reset_index = df.reset_index(drop=True)
print("DataFrame with reset index:")
print(df_reset_index.head())
```

```python
# 15
# to check whether the ‘Stadium’ column is present in your DataFrame or not.

# Check if 'Stadium' column is present
is_stadium_present = 'Stadium' in df.columns
print(f"Is 'Stadium' column present in the DataFrame? \n{is_stadium_present}")
```

```python
# 16
#  to convert the datatype of the ‘AwayTeamGoals’. (int to float)

# Convert the datatype of 'AwayTeamGoals' from int to float
df['AwayTeamGoals'] = df['AwayTeamGoals'].astype(float)
print("Data type of 'AwayTeamGoals' after conversion:")
print(df['AwayTeamGoals'].dtype)
```

```python
# 17
# to remove the last 10 rows from your DataFrame.

# Remove the last 10 rows from the DataFrame
df_removed_last_10 = df.iloc[:-10]
print("DataFrame after removing the last 10 rows:")
print(df_removed_last_10.tail())

```

```python
# 18
# to iterate over rows in your DataFrame.

# Iterate over rows in the DataFrame
print("Iterating over DataFrame rows:")
for index, row in df.iterrows():
    print(f"Index: {index}, HomeTeam: {row['HomeTeam']}, AwayTeam: {row['AwayTeam']}")
```

```python
# 19
#  to change the order of your DataFrame columns.

# Change the order of DataFrame columns
df_reordered = df[['AwayTeam', 'HomeTeam', 'HomeTeamGoals', 'AwayTeamGoals', 'Attendance', 'MatchType']]
print("DataFrame with reordered columns:")
print(df_reordered.head())
```

```python
# 20
# to delete DataFrame row(s) whose value is 0 in the ‘HomeTeamGoals’ columns.

# Delete rows where 'HomeTeamGoals' is 0
df_no_zero_home_goals = df[df['HomeTeamGoals'] != 0]
print("DataFrame after deleting rows with 'HomeTeamGoals' equal to 0:")
print(df_no_zero_home_goals)
```
