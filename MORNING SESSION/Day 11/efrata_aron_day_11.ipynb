```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv( r'heart.csv')
print(df)
print(df.info())
print(df.head())
```

```python
#check for missing values
print(df.isnull().sum())
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
print(df)
```

```python
# Check for missing values in each column
print("Missing values in each column:\n", df.isna().sum())
```

```python
# Explanatory Data Analysis(EDA)
print(df.describe())
```

```python
# Mean and Median
print("Mean:\n", df.mean())
print()
print("Median:\n", df.median())
print()
print("Mode:\n", df.mode().iloc[0])
```

```python
# Distributions of each column
df.hist(figsize=(12, 10))
plt.show()
```

```python
import seaborn as sns
# Box plot to identify outliers
plt.figure(figsize=(15, 12))
sns.boxplot(data=df, orient="h")
plt.show()
```

```python
# Scatter plot to examine relationships between age and trestbps

sns.scatterplot(x='age', y='trestbps', data=df)
plt.title('Scatter Plot of Age vs. trestbps')
plt.xlabel('age')
plt.ylabel('trestbps')

plt.show()
```

```python
sns.pairplot(df, vars=['age', 'trestbps', 'ca'])
plt.figure(figsize = (15, 10))
plt.show()
```

```python
# Skewness and Kurtosis
print("Skewness:\n", df.skew())
print("Kurtosis:\n", df.kurtosis())
```

```python
print("Range:\n", df.max() - df.min())
print()
print("Variance:\n", df.var())
print()
print("Standard Deviation:\n", df.std())
```
